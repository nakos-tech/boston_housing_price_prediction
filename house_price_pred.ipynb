```python
# data = fetch_california_housing()
# california_df = pd.DataFrame(data.data, columns=data.feature_names)

# print(california_df.head())
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston = pd.read_csv(url)

# Rename the 'medv' column to 'Price'
boston.rename(columns={'medv': 'Price'}, inplace=True)

# View the data
print(boston["Price"].head())
boston.head()
```

```python
# visualizing using seaborn also including a linear regression line 
#  sns.regplot() is a Seaborn function that creates a scatter plot with a regression line

import seaborn as sns

#  create a figure with two subplots side by side 
fig, axes = plt.subplots(1,2, figsize=(12, 5))

# plot 1
sns.regplot(x="rm", y="Price", data=boston, ax=axes[0])
axes[0].set_title("average number of rooms vs Price")

sns.regplot(x="lstat", y="Price", data=boston, ax=axes[1])
axes[1].set_title("LSTAT vs Price")

# Add a tight layout so plots don't overlap
plt.tight_layout()
plt.show()
```

```python
 #  we devide our data into independent and dependant variables 

# X = boston.iloc[:, :-1]
# y = boston.iloc[:, -1]

X = boston.drop("Price", axis=1)
y = boston["Price"]
```

```python
#  spliting the dataset into training and testing dataset

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=42)

print(X_train)
```

```python
# standadarzing the dataset 

from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()  # <-- This was missing!

# Fit and transform the training data
X_train = scaler.fit_transform(X_train)  

# Transform the test data (using the same scaler)
X_test = scaler.transform(X_test)  

print(X_train)
```

```python
#  we are going to use linear regression for our training

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

# print the coefficients and intercepts 

print(model.coef_)
print(model.intercept_)

# on which parameters the model has been trained 
model.get_params()
```

```python
#  use the trained model to predict values 

model_pred = model.predict(X_test)
print(model_pred)
```

```python
#  lets plot a scatter plot for the prediction
# Each point on the plot represents a single prediction.

# The x-coordinate is the actual value (y_test[i]) and the y-coordinate is the predicted value (model_pred[i]).
# If the model is perfect, all the points would lie on a 45-degree line (where actual = predicted).


plt.scatter(y_test, model_pred)

# calculating the errors 
pred_error = y_test - model_pred 
print(pred_error)

# plot the errors 
sns.displot(pred_error, kind="kde")
```

```python
#  performances metrics 
# we are measuring the performance of our model

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error, r2_score

mse = mean_squared_error(y_test, model_pred)
mae = mean_absolute_error(y_test, model_pred)

r2 = r2_score(y_test, model_pred)

print(mae, mse)
```
